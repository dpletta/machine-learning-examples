```python
import numpy as np
import matplotlib.pyplot as plt

def gibbs_sampler_bivariate_normal(initial_values, num_samples, correlation_coefficient):
    """
    Gibbs sampler for a bivariate normal distribution with specified correlation coefficient.

    Parameters:
    - initial_values: Initial values for the two variables.
    - num_samples: Number of samples to generate.
    - correlation_coefficient: Correlation coefficient between the two variables.

    Returns:
    - samples: Array containing the generated samples.
    """

    samples = np.zeros((num_samples, 2))

    current_values = np.array(initial_values)

    for i in range(num_samples):
        # Sample from the conditional distribution of the first variable
        mean_1 = 2 + correlation_coefficient * (current_values[1] - 3)
        current_values[0] = np.random.normal(mean_1, 1)

        # Sample from the conditional distribution of the second variable
        mean_2 = 3 + correlation_coefficient * (current_values[0] - 2)
        current_values[1] = np.random.normal(mean_2, 1)

        # Save the generated samples
        samples[i] = current_values

    return samples

# Set random seed for reproducibility
np.random.seed(42)

# Gibbs sampler parameters
initial_values = [0, 0]
num_samples = 2000
correlation_coefficient = 0.8

# Generate samples using the Gibbs sampler
samples = gibbs_sampler_bivariate_normal(initial_values, num_samples, correlation_coefficient)

# Scatter plot of the resulting samples
plt.scatter(samples[:, 0], samples[:, 1], color='blue', marker='o', label='Gibbs Samples')

# Add labels and title
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('Scatter Plot of Gibbs Sampler Samples')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```
