```python
import pandas as pd
import numpy as np
```

```python
# Set seed for reproducibility
np.random.seed(42)

# Create a 6x5 DataFrame with random integers between 0 and 10
data = np.random.randint(0, 11, size=(6, 5))
df = pd.DataFrame(data, columns=['Column1', 'Column2', 'Column3', 'Column4', 'Column5'])

# Display the DataFrame
print(df)
```

```python
# Display the DataFrame
print("DataFrame:")
print(df)
print()

# Print a specific value using label-based indexing (.at[])
row_label = 2
column_name = 'Column3'
value_at_label = df.at[row_label, column_name]
print(f"Value at row {row_label} and column '{column_name}': {value_at_label}")
print()

# Print a specific value using position-based indexing (.iat[])
row_index = 3
column_index = 1
value_at_position = df.iat[row_index, column_index]
print(f"Value at row {row_index} and column {column_index}: {value_at_position}")
```
